```python
# -----------------
# Operating profit:
# -----------------
"""
EBITDA is defined in its literal formula as:
    EBITDA = Net Profit + Interest + Taxes + Depreciation + Amortization
    
In its simplest form, however, EBITDA is defined as the following:
    EBITDA = Operating Profit + Depreciatiaon Expense + Amortization Expense
    
Operating profit is defined as 'ebit', next to 'ebitda', in the dataframe.
"""
dfw['ebit']
```

```python
# -----------------
# Define variables:
# -----------------
X = dfw[dfw.columns[~dfw.columns.isin(['ebit'])]]    # all other variables except ebit, independent variables (predictor)
Y = dfw[dfw.columns[dfw.columns.isin(['ebit'])]]     # ebit as the dependent (target)
```

```python
# --------------------
# Stepwise Regression:
# --------------------
"""
Perform a forward-backward feature selection 
    based on p-value from statsmodels.api
    Arguments:
        X                = all other variables except ebit, independent variables (predictor)
        Y                = ebit as the dependent (target)
        Variable_List    = list of variables with column names of X
        No_Reject        = include a variable if its p-value < No_Reject
        Reject           = exclude a variable if its p-value > Reject
        Add_Drop         = print the sequence of inclusions and exclusions
    Returns: list of selected variables
    Always set No_Reject < Reject to avoid infinite looping.
"""

def stepwise_regression(X, Y, 
                       Variable_List = [], 
                       No_Reject     = 0.01, 
                       Reject        = 0.05, 
                       Add_Drop      = True):
    included = list(Variable_List)
    while True:
        changed = False
        # Forward step
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index = excluded)
        for new_column in excluded:
            model = sm.OLS(Y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < No_Reject:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            if Add_Drop:
                print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval))
        
        # Backward step
        model = sm.OLS(Y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        # Use all coefficients except intercept
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() # Null if pvalues is empty
        if worst_pval > Reject:
            changed = True
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            if Add_Drop:
                print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included

result = stepwise_regression(X, Y)
```
